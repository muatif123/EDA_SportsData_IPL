## Exploratory Data Analysis

Exploratory Data Analysis(EDA) is essentially a type of storytelling for statisticians. It is a way of visualizing, summarizing and interpreting the data that is hidden in rows and column format. Once EDA is complete and insights are drawn, its feature can be used for supervised and unsupervised machine learning modelling.

### Sports Dataset

Sports dataset consists of IPL data from the year 2008-2019. The data is in the form of two csv files, 1st the matches played and 2nd is the individual player data of each team.

```python
# Importing the required libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
# Importing the dataset
df_matches = pd.read_csv('matches.csv')
df_matches
```

```python
%%time
df_deliveries = pd.read_csv('deliveries.csv')
df_deliveries.head()
```

```python
df_deliveries.shape
```

```python
# Dropping the unrequired columns
df_matches = df_matches.drop(columns = 'umpire3')
df_matches.head()
```

##### Calculating the wins by each team in each season

```python
wins = df_matches.groupby('season')['winner'].value_counts()
wins
```

##### Number of matches played in each season

```python
plt.figure(figsize = (18,10))
sns.countplot('season', data = df_matches, palette = "winter")
plt.title("Number of Matches played in each IPL season", fontsize = 20)
plt.xlabel("Season", fontsize = 15)
plt.ylabel('Matches', fontsize = 15)
plt.show()
```

##### Number of matches won by each team

```python
plt.figure(figsize = (18,10))
sns.countplot(x = 'winner',data = df_matches, palette = 'cool')
plt.title("Numbers of matches won by team ", fontsize = 20)
plt.xticks(rotation = 60)
plt.xlabel("Teams",fontsize = 15)
plt.ylabel("Number of wins",fontsize = 15)
plt.show()
```

##### Pie Chart for Match Result

```python
df_matches['win_by'] = np.where(df_matches['win_by_runs'] > 0, 'Bat-First', 'Bowl-First')
```

```python
win = df_matches.win_by.value_counts()
labels = np.array(win.index)
sizes = win.values
colors = ['#FFBF00', '#FA8072']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels = labels, colors = colors,
        autopct = '%1.1f%%', shadow = True,startangle = 90)
plt.title('Match Result',fontsize = 20)
plt.axis('Equal',fontsize = 10)
plt.show()
```

##### Numbers of matches won by batting and bowling first 

```python
plt.figure(figsize = (18,10))
sns.countplot('season', hue = 'win_by',data = df_matches, palette = 'hsv')
plt.title("Numbers of matches won by batting and bowling first ",fontsize = 20)
plt.xlabel("Season",fontsize = 15)
plt.ylabel("Count",fontsize = 15)
plt.show()
```

##### Pie Chart for Toss result

```python
toss = df_matches.toss_decision.value_counts()
labels = np.array(toss.index)
sizes = toss.values
colors = ['#FFBF00', '#FA8072']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels = labels, colors = colors,
        autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.title('Toss result', fontsize = 20)
plt.axis('equal',fontsize = 10)
plt.show()
```

##### Numbers of matches won by Toss result in each season

```python
plt.figure(figsize = (18,10))
sns.countplot('season',hue = 'toss_decision',data = df_matches, palette = 'afmhot')
plt.title("Numbers of matches won by Toss result ",fontsize = 20)
plt.xlabel("Season",fontsize = 15)
plt.ylabel("Count",fontsize = 15)
plt.show()
```

##### Displaying the winner of each season

```python
final_matches = df_matches.drop_duplicates(subset = ['season'], keep = 'last')

final_matches[['season','winner']].reset_index(drop = True).sort_values('season')
```

##### Top player based on match Winners

```python
plt.figure(figsize = (18, 10))
top_players = df_matches.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0, 20])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient = 'v', palette = "hsv");
plt.show()
```

##### Numbers of fours hit by team

```python
four_data = df_deliveries[df_deliveries['batsman_runs'] == 4]
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

##### Numbers of fours hit by players

```python
batsman_four = four_data.groupby('batsman')['batsman_runs'].agg([('four', 'count')]).reset_index().sort_values('four', ascending=0)
ax = batsman_four.iloc[:10,:].plot('batsman', 'four', kind = 'bar', color = 'blue')
plt.title("Numbers of fours hit by playes ",fontsize = 20)
plt.xticks(rotation = 50)
plt.xlabel("Player name",fontsize = 15)
plt.ylabel("No of fours",fontsize = 15)
plt.show()
```

##### Numbers of sixes hit by each team

```python
six_data = df_deliveries[df_deliveries['batsman_runs'] == 6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six', 'sum'),('sixes', 'count')])
```

##### Numbers of sixes hit by players

```python
batsman_six = six_data.groupby('batsman')['batsman_runs'].agg([('six', 'count')]).reset_index().sort_values('six',ascending=0)
ax = batsman_six.iloc[:10,:].plot('batsman', 'six',kind = 'bar',color = 'red')
plt.title("Numbers of six hit by playes ",fontsize = 20)
plt.xticks(rotation = 50)
plt.xlabel("Player name",fontsize = 15)
plt.ylabel("No of six",fontsize = 15)
plt.show()
```

##### Top 10 leading run scorer in IPL

```python
batsman_score = df_deliveries.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum', ascending = False).reset_index(drop = True)
batsman_score = batsman_score.rename(columns = {'sum': 'batsman_runs'})
print("Top 10 Leading Run Scorer in IPL")
batsman_score.iloc[:10, :]
```

##### Number of matches played by each batsman

```python
matches_player = df_deliveries[["match_id", "player_dismissed"]]
matches_player = matches_player.groupby("player_dismissed")["match_id"].count().reset_index().sort_values(by = "match_id", ascending = False).reset_index(drop = True)
matches_player.columns = ["batsman", "No_of Matches"]
matches_player.head(5)
```

##### Dismissals in IPL

```python
plt.figure(figsize=(18,10))
ax=sns.countplot(df_deliveries.dismissal_kind)
plt.title("Dismissals in IPL", fontsize = 20)
plt.xlabel("Dismissals kind", fontsize = 15)
plt.ylabel("count", fontsize = 15)
plt.xticks(rotation = 90, fontsize = 15)
plt.show()
```
